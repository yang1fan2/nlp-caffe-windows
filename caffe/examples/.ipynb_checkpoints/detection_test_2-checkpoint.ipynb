```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import os
import time
import sys
sys.path.append('../python')
import caffe
```

```python
pretrained_model = '../models/bvlc_reference_rcnn_ilsvrc13/bvlc_reference_rcnn_ilsvrc13.caffemodel'
model_def = '../models/bvlc_reference_rcnn_ilsvrc13/deploy.prototxt'
gpu = 1
raw_scale=255
input_scale = None
channel_swap = [2, 1, 0]
context_pad = 16
mean_file = '../python/caffe/imagenet/ilsvrc_2012_mean.npy'
mean = np.load(mean_file)
#print mean
```

```python
# Make detector.
detector = caffe.Detector(model_def, pretrained_model,
        gpu, mean,
        input_scale, raw_scale,
        channel_swap,
        context_pad)
```

```python
images_windows = [('images/fish-bike.jpg', np.array([[9, 230, 150, 330]]))]
                  #, ('images/cat.jpg', np.array([[150, 1, 350, 350]]))]
detections = detector.detect_windows(images_windows)
for i in range(len(detections)):
    print detections[i]
```

```python
with open('../data/ilsvrc12/det_synset_words.txt') as f:
    labels_df = pd.DataFrame([
        {
            'synset_id': l.strip().split(' ')[0],
            'name': ' '.join(l.strip().split(' ')[1:]).split(',')[0]
        }
        for l in f.readlines()
    ])
labels_df.sort('synset_id');
```

```python
predictions_df = pd.DataFrame(np.array([detections[0]['prediction']]), columns=labels_df['name'])
print(predictions_df.iloc[0])
```

```python
max_s = predictions_df.max(0)
max_s.sort(ascending=False)
print(max_s[:10])
```
